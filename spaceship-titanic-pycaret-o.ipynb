## 👩‍💻 User Modules
สร้างโมดูลเพื่อใช้ในการ visualization

```python
def multi_table(table_list):
    return HTML(
        f"<table><tr> {''.join(['<td>' + table._repr_html_() + '</td>' for table in table_list])} </tr></table>")
```

```python
def cat_dist(data, var, hue, msg_show=True):
    total_cnt = data[var].count()
    f, ax = plt.subplots(1, 2, figsize=(25, 8))
    hues = [None, hue]
    titles = [f"{var}'s distribution", f"{var}'s distribution by {hue}"]

    for i in range(2):
        sns.countplot(data[var], edgecolor='black', hue=hues[i], linewidth=1, ax=ax[i], data=data)
        ax[i].set_xlabel(var, weight='bold', size=13)
        ax[i].set_ylabel('Count', weight='bold', size=13)
        ax[i].set_facecolor(BACKCOLOR)
        ax[i].spines[['top', 'right']].set_visible(False)
        ax[i].set_title(titles[i], size=15, weight='bold')
        for patch in ax[i].patches:
            x, height, width = patch.get_x(), patch.get_height(), patch.get_width()
            if msg_show:
                ax[i].text(x + width / 2, height + 3, f'{height} \n({height / total_cnt * 100:2.2f}%)', va='center', ha='center', size=12, bbox={'facecolor': 'white', 'boxstyle': 'round'})
    plt.show()
```

```python
def continuous_dist(data, x, y):
    f, ax = plt.subplots(1, 4, figsize=(35, 10))
    sns.histplot(data=train, x=y, hue=x, ax=ax[0], element='step')
    sns.violinplot(x=data[x], y=data[y], ax=ax[1], edgecolor='black', linewidth=1)
    sns.boxplot(x=data[x], y=data[y], ax=ax[2])
    sns.stripplot(x=data[x], y=data[y], ax=ax[3])
    for i in range(4):
        ax[i].spines[['top','right']].set_visible(False)
        ax[i].set_xlabel(x, weight='bold', size=20)
        ax[i].set_ylabel(y, weight='bold', size=20)
        ax[i].set_facecolor(BACKCOLOR)
    f.suptitle(f"{y}'s distribution by {x}", weight='bold', size=25)
    plt.show()
```

### Cabin(Nominal)  
Cabin คือหมายเลขห้องผู้โดยสารซึ่งประกอบด้วยหลายประเภท โดยบางห้องชื่อเหมือนกับชื่อของห้องผู้โดยสาร โดยคำแรกและคำสุดท้ายเป็นประเภทของห้องโดยสาร

```python
tmp = train.copy()
tmp['Deck'] = train.Cabin.apply(lambda x:str(x)[:1])
tmp['side'] = train.Cabin.apply(lambda x:str(x)[-1:])

cat_dist(tmp, var='Deck', hue='Transported')
```

```python
cat_dist(tmp, var='side', hue='Transported')
```

## 👩‍💻Missing Value
แต่ละ column มี Missing Value โดยเรามีวิธีจัดการกับ Missing Value ตามลักษณะของข้อมูลดังนี้
> **1. Categorical Variables** (CryoSleep, Cabin, VIP, HomePlanet, Destination)  
> **2. Continuous Variables** or numerical (Age , RoomService, FoodCourt, ShoppingMall, Spa, VRDeck)

```python
# Columns with missing values
na_cols = all_data.columns[all_data.isna().any()].tolist()

# Missing values summary
missing_values = pd.DataFrame(all_data[na_cols].isna().sum(), columns=['Number_missing'])
missing_values['Percentage_missing']=np.round(100*missing_values['Number_missing']/len(all_data),2)
missing_values
```

```python
# เติมค่าใช้จ่ายต่างๆ ด้วย 0
all_data['RoomService']=all_data['RoomService'].fillna(0)
all_data['FoodCourt']=all_data['FoodCourt'].fillna(0)
all_data['ShoppingMall']=all_data['ShoppingMall'].fillna(0)
all_data['Spa']=all_data['Spa'].fillna(0)
all_data['VRDeck']=all_data['VRDeck'].fillna(0)

# เติม name ด้วยค่าของตัวข้างบน
all_data['Name']=all_data['Name'].fillna(method='ffill')

Age_cols = ['Age']
categoriacl_cols = ['HomePlanet','Destination','Cabin','VIP','CryoSleep']

# เติมอายุด้วย Mean
# เติม categoriacl values ที่เหลือด้วยค่า mode
def fill_missing(df):
    for ft in categoriacl_cols:
        df[ft].fillna(df[ft].mode()[0],inplace=True)
       
    for ft in Age_cols:
        df[ft].fillna(df[ft].mean(),inplace=True)    
    return df

fill_missing(all_data)
```


<a id="5.2"></a>
## 👩‍💻Create new vaireble
> * **PassengerId** - A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. People in a group are often family members, but not always.
> * **Cabin** - The cabin number where the passenger is staying. Takes the form deck/num/side, where side can be either P for Port or S for Starboard.

จาก **Cabin** เป็นรหัสที่สามารถสื่อได้ถึง deck/num/side โดยเราจะแยกเป็น 3 คอลัมน์ใหม่ได้แก่ 
> 1. Cabin_Deck
> 2. Cabin_Num - เราจะไม่ใช้คอลัมน์นี้
> 3. Cabin_Side - P (Port) or S (Starboard)

จาก **PassengerId(gggg_pp)** เราทราบได้ว่าใครมาจากกลุ่มเดียวกันจากรหัสสี่ตัวแรก (gggg) และรหัสสองตัวหลัง (pp) _ เป็นตัวเลขประจำตัวผู้โดยสารในกลุ่มนั้นๆ ดังนั้นเราจะสร้างเป็นคอลัมน์ใหม่ 2 คอลัมน์จากคอลัมน์ PassengerId จะได้คอลัมน์ใหม่เป็น 
> 1. PassengerGroup - gggg 
> 2. PassengerNo -  pp

จาก **ค่าใช้จ่ายทั้งหมด** - 'RoomService'+'FoodCourt'+'ShoppingMall'+'Spa'+'VRDeck' นำมาสร้างเป็นคอลัมน์ใหม่ **TotalSpend**

จาก **Age** เราจะแบ่งอายุผู้โดยสารเป็น 2 กลุ่มได้แก่
> True --> Adult - ผู้โดยสารที่อายุมากกว่า 18 ปี\
> False --> Not Adult - ผู้โดยสารที่อายุน้อยกว่า 18 ปี

```python
all_data['Cabin_Deck'] = all_data.Cabin.apply(lambda x:str(x)[:1])
all_data['Cabin_Side'] = all_data.Cabin.apply(lambda x:str(x)[-1:])

all_data['PassengerGroup'] = all_data['PassengerId'].apply(lambda x: x.split('_')[0])
all_data['PassengerNo'] = all_data['PassengerId'].apply(lambda x: x.split('_')[1])

all_data['TotalSpend'] = all_data['RoomService'] + all_data['FoodCourt'] + all_data['ShoppingMall'] + all_data['Spa'] + all_data['VRDeck']

all_data['Adult']=True
all_data.loc[all_data['Age']<18, 'Adult']=False
```

### ☂️ Setup 
สิ่งแรกที่ขาดไม่ได้เลยสำหรับการใช้งาน Pycaret ก็คือการ Setup ข้อมูล หรือที่เราเรียกกันว่าการเตรียมข้อมูล ซึ่งถือว่าเป็นหนึ่งในสิ่งมหัศจรรย์ของ Pycaret เลยก็ว่าได้ เพราะแค่กำหนดข้อมูลตั้งต้นคู่กับชื่อคอลัมน์ที่เราต้องการทำนาย ก็สามารถนำไปใส่โมเดลเพื่อทำการทดลองได้แล้ว ซึ่งจะใช้คำสั่งง่าย ๆ ที่ว่า setup

โดยเรามีการ setup ค่าต่าง ๆ ดังนี้ : 
 1. data = ข้อมูลที่ต้องการดำเนินการ
 2. target = ชื่อคอลัมน์ที่เป็นสิ่งที่โมเดลต้องการทำนาย ซึ่งอาจเป็น binary หรือ multiclass ก็ได้ 
 3. session_id = ชื่อการกำหนดให้การสุ่มข้อมูลนั้นเป็นชุดข้อมูลเดิมทุกครั้ง คล้ายกับ 'random_state' ใน scikit-learn
 4. train_size = สัดส่วนของชุดข้อมูลที่จะใช้สำหรับการ train 
 5. fold_strategy = ตัวเลือกในการทำ cross validation strategy โดยเลือกใช้ ‘stratifiedkfold’ (default) 
 6. fold = จำนวนจุดทศนิยมที่เราจะแสดงค่าในการทำ cross validation อย่างน้อยต้อง 2 ตำแหน่ง (default = 10)
 7. fold_shuffle = เป็นการควบคุมการสับเปลี่ยน parameter ของ CV 
 8. silent = เป็นการควบคุมการแสดงผล data type ของข้อมูลเมื่อเกิดการ run code
 9. ignore_low_variance = เป็นการตั้งค่าว่าต้องการให้ไม่แสดงผลลัพธ์ที่มีความแปรปรวนต่ำหรือไม่ (When set to True, all categorical features with insignificant variances are removed from the data)
 10. remove_multicollinearity = When set to True, perfect collinearity (features with correlation = 1) is removed from the dataset, when two features are 100% correlated, one of it is randomly removed from the dataset (หากข้อมูลมีความสอดคล้องกัน 100% (ซ้ำกัน) ให้ดำนินการลบข้อมูลหนึ่งออก)
 11. normalize = เมื่อตั้งค่าเป็น True จะแปลงคุณสมบัติตัวเลขโดยปรับขนาดให้เป็นช่วงที่กำหนด
 12. normalize_method = กำหนดวิธีการปรับขนาดให้เป็นค่ามาตรฐาน โดยมีค่าเริ่มต้น คือ 'zscore' 
     * robust: scales and translates each feature according to the Interquartile range. When the dataset contains outliers, robust scaler often gives better results.
