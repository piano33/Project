#Association Rule Mining - ชุดข้อมูลระดับความรุนแรง DIE

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
url = 'https://raw.githubusercontent.com/rawisara01/DSI314/main/Association%20Rule%20Mining/level_acc_DIE.csv'
df_DIE = pd.read_csv(url)
```

```python
# ลบคอลัมน์ที่ไม่ได้ใช้
df_DIE.drop(columns=['Unnamed: 0', 'day_acc', 'years_acc', 'month_acc'], inplace=True, axis=1)
df_DIE
```

```python
# download ข้อมูลเพื่อนำไปทำ dashboard
from google.colab import files
df_DIE.to_excel('level_DIE_Visual.xlsx') 
files.download('level_DIE_Visual.xlsx')
```

```python
# ลบคอลัมน์ที่ไม่ใช้งาน
df_DIE.drop('level_acc', inplace=True, axis=1)
```

```python
df_DIE['products_DIE'] = df_DIE[df_DIE.columns[0:]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1)
```

```python
df_DIE.drop(['route_acc', 'type_acc', 'cnd_acc',  'scene_acc', 'road_surface', 'time_acc'], axis=1, inplace=True)
```

```python
df_DIE
```

```python
data = list(df_DIE["products_DIE"].apply(lambda x:x.split(" , ") ))
```

```python
data = list(df_DIE["products_DIE"].apply(lambda x:x.split(",") ))
data
```

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder

a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df_DIE1 = pd.DataFrame(a_data,columns=a.columns_)
df_DIE1 = df_DIE1.replace(False,0)
df_DIE1
```

```python
df_DIE1 = apriori(df_DIE1, min_support = 0.2, use_colnames = True,)
df_DIE1
```

```python
#Associan rule function
df_DIE1_ar = association_rules(df_DIE1, metric = "lift", min_threshold = 0.6)
df_DIE1_ar
```

```python
df_DIE2_ar = df_DIE1_ar.query(" lift > 1.15 ")
```
