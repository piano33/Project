#Association Rule Mining - ชุดข้อมูลระดับความรุนแรง INJ

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
url = 'https://raw.githubusercontent.com/rawisara01/DSI314/main/Association%20Rule%20Mining/level_acc_INJ.csv'
df_INJ =pd.read_csv(url)
```

```python
# ลบคอลัมน์ที่ไม่ได้ใช้
df_INJ.drop(columns=['Unnamed: 0','day_acc', 'years_acc', 'month_acc'], inplace=True, axis=1)
df_INJ
```

```python
# download ข้อมูลเพื่อนำไปทำ dashboard
from google.colab import files
df_INJ.to_excel('level_INJ_Visual.xlsx') 
files.download('level_INJ_Visual.xlsx')
```

```python
# ลบคอลัมน์ที่ไม่ใช้งาน
df_INJ.drop('level_acc', inplace=True, axis=1)
```

```python
df_INJ['products_INJ'] = df_INJ[df_INJ.columns[0:]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
```

```python
df_INJ.drop(['route_acc', 'type_acc', 'cnd_acc', 'scene_acc', 'road_surface', 'time_acc'], axis=1, inplace=True)
```

```python
df_INJ
```

```python
data= list(df_INJ["products_INJ"].apply(lambda x:x.split(" , ") ))
```

```python
data = list(df_INJ["products_INJ"].apply(lambda x:x.split(",") ))
data
```

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder

a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df_INJ1 = pd.DataFrame(a_data,columns=a.columns_)
df_INJ1 = df_INJ1.replace(False,0)
df_INJ1
```

```python
df_INJ1 = apriori(df_INJ1, min_support = 0.2, use_colnames = True,)
df_INJ1
```

```python
#Associan rule function
df_INJ1_ar = association_rules(df_INJ1, metric = "confidence", min_threshold = 0.6)
df_INJ1_ar
```

```python
df_INJ2_ar = df_INJ1_ar.query(" lift > 1.089 ")
```
